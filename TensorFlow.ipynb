```python
def   c_to_f(c):
  return c * 1.8 +32
```

```python
Cs=[0,1,4,12,30,45,100,105,90,77,-40,-10,8,15,22,38]
Fs=[]

for i in Cs:
  f =c_to_f(i)
  Fs.append(f)
```

```python
Fs
```

```python
import tensorflow as tf
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(loss='mse' , optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))
```

```python
import numpy as np

Cs_np = np.array(Cs)
Fs_np = np.array(Fs)

history = model.fit(Cs_np , Fs_np, epochs=600)
```

```python
import numpy as np
c= 101.65

print(f"model prediction = " , model.predict(np.array([c])))
print(f"actual value = " , c_to_f(c))
```
