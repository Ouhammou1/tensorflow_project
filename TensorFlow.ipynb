```python
def clean(path):
    images = os.listdir(path)
    for image in images:
        if os.path.getsize(os.path.join(path , image)) == 0 or image.split('.')[len(image.split('.')) -1] != 'jpg':
            images.remove(image)
    return images

```

```python
import os
import shutil

WORKING_DIR = "/kaggle/working/PetImages"

if not os.path.exists(WORKING_DIR):
    shutil.copytree("/kaggle/input/data-cat-dig/PetImages", WORKING_DIR)

data_path = WORKING_DIR

def train_validation_dirs(dir_name, data, SPLIT_SIZE=0.7, data_path=data_path, data_length=7000):
    training_length = round(SPLIT_SIZE * data_length)
    
    train_set = data[0:training_length]
    val_set = data[training_length:data_length]

    os.makedirs(os.path.join(data_path, 'training', dir_name), exist_ok=True)
    os.makedirs(os.path.join(data_path, 'validation', dir_name), exist_ok=True)

    for image in train_set:
        shutil.copyfile(
            os.path.join(data_path, dir_name, image),
            os.path.join(data_path, 'training', dir_name, image)
        )
            
    for image in val_set:
        shutil.copyfile(
            os.path.join(data_path, dir_name, image),
            os.path.join(data_path, 'validation', dir_name, image)
        )

```

```python
train_validation_dirs('Dog' , dogs)
train_validation_dirs('Cat' , dogs)



folder_path = "/kaggle/working/PetImages/training/Dog"
folder_path1 = "/kaggle/working/PetImages/training/Cat"


if os.path.exists(folder_path) and os.path.exists(folder_path1):
    print("exist ")
else:
    print("not exist ")

print(os.listdir("/kaggle/working/PetImages/training"))
print(os.listdir("/kaggle/working/PetImages/validation"))
```
